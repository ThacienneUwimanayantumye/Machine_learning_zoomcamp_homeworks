# Homework 6: Decision Trees and Random Forests

This notebook contains solutions to all questions from Homework 6, focusing on decision trees, random forests, and XGBoost for predicting car fuel efficiency.

## Dataset
We'll be using the car fuel efficiency dataset downloadable [here](https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv).


```python
# importing all necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
```

```python
# Download the dataset
import urllib.request

url = 'https://raw.githubusercontent.com/alexeygrigorev/datasets/master/car_fuel_efficiency.csv'
df = pd.read_csv(url)

print(f"Dataset shape: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
print(f"\nTotal missing values: {df.isnull().sum().sum()}")
```

```python
# Data preparation
# Fill missing values with zeros
df = df.fillna(0)

# Separate features and target
y = df['fuel_efficiency_mpg'].values
del df['fuel_efficiency_mpg']

# Train/validation/test split: 60%/20%/20%
df_train_full, df_test, y_train_full, y_test = train_test_split(
    df, y, test_size=0.2, random_state=1
)

df_train, df_val, y_train, y_val = train_test_split(
    df_train_full, y_train_full, test_size=0.25, random_state=1  # 0.25 of 80% = 20%
)

print(f"Training set: {df_train.shape}")
print(f"Validation set: {df_val.shape}")
print(f"Test set: {df_test.shape}")
```

```python
# Use DictVectorizer to turn dataframes into matrices
dv = DictVectorizer(sparse=True)

# Convert to dictionaries
train_dict = df_train.to_dict(orient='records')
val_dict = df_val.to_dict(orient='records')
test_dict = df_test.to_dict(orient='records')

# Transform
X_train = dv.fit_transform(train_dict)
X_val = dv.transform(val_dict)
X_test = dv.transform(test_dict)

print(f"Feature matrix shape - Training: {X_train.shape}, Validation: {X_val.shape}, Test: {X_test.shape}")
print(f"Feature names: {dv.get_feature_names_out()}")
```

## Question 1: Decision Tree with max_depth=1

Let's train a decision tree regressor to predict the `fuel_efficiency_mpg` variable with `max_depth=1` and identify which feature is used for splitting the data.


```python
# Train decision tree with max_depth=1
dt = DecisionTreeRegressor(max_depth=1, random_state=1)
dt.fit(X_train, y_train)

# Get feature names
feature_names = dv.get_feature_names_out()

# Find the feature used for splitting (first split)
tree = dt.tree_
feature_used = tree.feature[0]
feature_name = feature_names[feature_used]

print(f"Feature used for splitting: {feature_name}")
print(f"\nDecision tree structure:")
print(f"Root node uses feature: {feature_name}")
print(f"Threshold: {tree.threshold[0]:.4f}")
print(f"Number of samples in root: {tree.n_node_samples[0]}")
```

```python
# Visualize the tree structure more clearly
from sklearn.tree import export_text

tree_rules = export_text(dt, feature_names=list(feature_names), max_depth=1)
print("Decision Tree Rules:")
print(tree_rules)
```

### Answer to Question 1

**Answer: `vehicle_weight`**

**Explanation**: 
The decision tree analysis clearly shows that `vehicle_weight` is the feature used for the root split. The tree splits at threshold 3022.11 lbs, creating two branches:
- Cars with weight â‰¤ 3022.11 lbs: Average fuel efficiency = 16.88 MPG
- Cars with weight > 3022.11 lbs: Average fuel efficiency = 12.94 MPG

This makes physical sense - heavier vehicles typically have lower fuel efficiency. The algorithm chose this feature because it provides the maximum information gain for predicting fuel efficiency, making it the most discriminative single feature in the dataset.

### Insights for Question 1

The decision tree with `max_depth=1` uses a single split to partition the data. The feature identified above is the most informative feature for predicting fuel efficiency at the root level, indicating its primary importance in the dataset. This result is consistent with Question 5's feature importance analysis, where `vehicle_weight` also dominates.


## Question 2: Random Forest with n_estimators=10

Train a random forest regressor with `n_estimators=10`, `random_state=1`, and calculate the RMSE on the validation data.


```python
# Train random forest with n_estimators=10
rf = RandomForestRegressor(n_estimators=10, random_state=1, n_jobs=-1)
rf.fit(X_train, y_train)

# Predict on validation set
y_pred = rf.predict(X_val)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_val, y_pred))

print(f"RMSE on validation data: {rmse:.4f}")
print(f"\nPrediction statistics:")
print(f"Mean predicted value: {y_pred.mean():.4f}")
print(f"Mean actual value: {y_val.mean():.4f}")
```

## Question 3: Finding Optimal n_estimators

Experiment with different values of `n_estimators` from 10 to 200 with step 10, and identify after which value the RMSE stops improving (considering 3 decimal places).


```python
# Try different values of n_estimators from 10 to 200 with step 10
n_estimators_range = range(10, 201, 10)
rmse_scores = []

for n_est in n_estimators_range:
    rf = RandomForestRegressor(n_estimators=n_est, random_state=1, n_jobs=-1)
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, y_pred))
    rmse_scores.append(rmse)
    
    if n_est % 50 == 0 or n_est <= 30:
        print(f"n_estimators={n_est:3d}, RMSE={rmse:.6f}")

# Find where RMSE stops improving (considering 3 decimal places)
rmse_rounded = [round(r, 3) for r in rmse_scores]
improvement_stopped = False
stopped_at = None

for i in range(1, len(rmse_rounded)):
    if rmse_rounded[i] >= rmse_rounded[i-1]:
        if not improvement_stopped:
            stopped_at = n_estimators_range[i]
            improvement_stopped = True
    else:
        improvement_stopped = False
        stopped_at = None

if stopped_at is None:
    stopped_at = n_estimators_range[-1]
    
print(f"\nRMSE stops improving after n_estimators = {stopped_at}")
```

```python
# Visualize the RMSE values
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(list(n_estimators_range), rmse_scores, marker='o', linewidth=2, markersize=4)
plt.xlabel('n_estimators', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.title('RMSE vs n_estimators (Random Forest)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(range(10, 201, 20))
plt.tight_layout()
plt.show()

# Show detailed results for key values
print("\nDetailed RMSE values (rounded to 3 decimal places):")
for i, (n_est, rmse) in enumerate(zip(n_estimators_range, rmse_scores)):
    if n_est % 20 == 0 or n_est <= 30:
        print(f"n_estimators={n_est:3d}, RMSE={rmse:.6f} ({rmse_rounded[i]:.3f})")
```

```python
# More precise check: when does improvement stop (3 decimal places)
print("Checking when RMSE stops improving (considering 3 decimal places):")
print("\nFinding the last n_estimators value where RMSE improves...")

# Find the last value where we see improvement (rounded to 3 decimal places)
last_improvement = None
for i in range(1, len(rmse_rounded)):
    prev_rmse = rmse_rounded[i-1]
    curr_rmse = rmse_rounded[i]
    
    if curr_rmse < prev_rmse:
        # Improvement found
        last_improvement = n_estimators_range[i]
        print(f"âœ“ n_estimators={n_estimators_range[i]:3d}: RMSE improved from {prev_rmse:.3f} to {curr_rmse:.3f}")
    else:
        # No improvement or got worse
        if curr_rmse > prev_rmse:
            print(f"âœ— n_estimators={n_estimators_range[i]:3d}: RMSE worsened from {prev_rmse:.3f} to {curr_rmse:.3f}")
        else:
            print(f"= n_estimators={n_estimators_range[i]:3d}: RMSE stayed at {curr_rmse:.3f}")

if last_improvement is None:
    # If never improved, use the first value
    last_improvement = n_estimators_range[0]
    print(f"\nNote: RMSE never improved, using first value")

print(f"\n" + "="*60)
print(f"Answer: RMSE stops improving after n_estimators = {last_improvement}")
print(f"This means the last improvement occurred at n_estimators = {last_improvement}")
print("="*60)
```

### Answer to Question 3

**Answer: 180**

**Explanation**:
Based on the detailed analysis above, the last improvement in RMSE (when rounded to 3 decimal places) occurred at **n_estimators = 180**, where RMSE improved from 0.443 to 0.442. 

After 180:
- n_estimators = 190: RMSE stayed at 0.442 (rounded)
- n_estimators = 200: RMSE stayed at 0.442 (rounded)

The question asks "after which value of n_estimators does RMSE stop improving?" Since the last improvement happened at 180, and subsequent values (190, 200) show no further improvement when rounded to 3 decimal places, the answer is **180**.

This makes sense from a machine learning perspective - we see diminishing returns as we add more trees. The model reaches a stable performance level where additional trees don't significantly reduce the error rate.

### Insights for Question 3

As the number of estimators increases, the model's performance generally improves, but the marginal gains diminish. ## Question 4: Finding Best max_depth

Try different values of `max_depth`: [10, 15, 20, 25]. For each value, try different `n_estimators` from 10 to 200 (step 10) and calculate the mean RMSE. Select the best `max_depth` based on mean RMSE.


```python
# Try different max_depth values
max_depth_values = [10, 15, 20, 25]
n_estimators_range = range(10, 201, 10)

plt.figure(figsize=(12, 6))
results = {}

for max_d in max_depth_values:
    rmse_scores = []
    print(f"\nTesting max_depth={max_d}")
    
    for n_est in n_estimators_range:
        rf = RandomForestRegressor(
            n_estimators=n_est, 
            max_depth=max_d, 
            random_state=1, 
            n_jobs=-1
        )
        rf.fit(X_train, y_train)
        y_pred = rf.predict(X_val)
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        rmse_scores.append(rmse)
    
    mean_rmse = np.mean(rmse_scores)
    results[max_d] = mean_rmse
    print(f"Mean RMSE for max_depth={max_d}: {mean_rmse:.6f}")
    plt.plot(list(n_estimators_range), rmse_scores, marker='o', label=f'max_depth={max_d}', linewidth=2, markersize=3)

plt.xlabel('n_estimators', fontsize=12)
plt.ylabel('RMSE', fontsize=12)
plt.title('RMSE vs n_estimators for Different max_depth Values', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.xticks(range(10, 201, 20))
plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("Summary of Mean RMSE for each max_depth:")
for max_d, mean_rmse in sorted(results.items()):
    print(f"max_depth={max_d:2d}: Mean RMSE = {mean_rmse:.6f}")

best_max_depth = min(results, key=results.get)
print(f"\nBest max_depth: {best_max_depth} (Mean RMSE = {results[best_max_depth]:.6f})")
```

### Answer to Question 4

**Answer: 10**

**Explanation**:
The results show the mean RMSE for each `max_depth` value:
- max_depth = 10: Mean RMSE = **0.441808** (lowest)
- max_depth = 15: Mean RMSE = 0.445417
- max_depth = 20: Mean RMSE = 0.446253
- max_depth = 25: Mean RMSE = 0.445910

The best (lowest) mean RMSE is achieved with **max_depth = 10**.

This is an interesting and important finding - shallower trees actually perform better than deeper ones! This demonstrates the bias-variance tradeoff in action. Deeper trees (15, 20, 25) are more complex and tend to overfit the training data, leading to worse generalization on the validation set. The max_depth=10 setting provides the right balance - it's deep enough to capture important patterns but shallow enough to maintain good generalization.

### Insights for Question 4

The mean RMSE across different `n_estimators` values helps identify the most robust `max_depth` setting. A lower mean RMSE indicates that a particular `max_depth` consistently performs well regardless of the number of trees, suggesting better generalization. The fact that max_depth=10 outperforms deeper trees is a key insight - it shows that limiting tree depth is an effective regularization technique for Random Forests on this dataset.

## Question 6: XGBoost with Different eta Values

Train XGBoost models with `eta=0.3` and `eta=0.1` (learning rate) and compare their RMSE scores on the validation dataset to determine which eta leads to better performance.


```python
# Install XGBoost if not already installed (uncomment if needed)
# !pip install xgboost

# Convert sparse matrices to DMatrix format for XGBoost
# XGBoost works better with dense matrices or DMatrix
dtrain = xgb.DMatrix(X_train, label=y_train, enable_categorical=False)
dval = xgb.DMatrix(X_val, label=y_val, enable_categorical=False)

# Create watchlist for monitoring
watchlist = [(dtrain, 'train'), (dval, 'val')]
```

```python
# Train model with eta=0.3
xgb_params_03 = {
    'eta': 0.3, 
    'max_depth': 6,
    'min_child_weight': 1,
    'objective': 'reg:squarederror',
    'nthread': 8,
    'seed': 1,
    'verbosity': 1,
}

print("Training XGBoost model with eta=0.3...")
model_03 = xgb.train(
    xgb_params_03,
    dtrain,
    num_boost_round=100,
    evals=watchlist,
    verbose_eval=False
)

# Predict on validation set
y_pred_03 = model_03.predict(dval)
rmse_03 = np.sqrt(mean_squared_error(y_val, y_pred_03))

print(f"\nRMSE with eta=0.3: {rmse_03:.6f}")
```

```python
# Train model with eta=0.1
xgb_params_01 = {
    'eta': 0.1, 
    'max_depth': 6,
    'min_child_weight': 1,
    'objective': 'reg:squarederror',
    'nthread': 8,
    'seed': 1,
    'verbosity': 1,
}

print("Training XGBoost model with eta=0.1...")
model_01 = xgb.train(
    xgb_params_01,
    dtrain,
    num_boost_round=100,
    evals=watchlist,
    verbose_eval=False
)

# Predict on validation set
y_pred_01 = model_01.predict(dval)
rmse_01 = np.sqrt(mean_squared_error(y_val, y_pred_01))

print(f"\nRMSE with eta=0.1: {rmse_01:.6f}")
```

```python
# Compare the results
print("\n" + "=" * 60)
print("Comparison of XGBoost models with different eta values:")
print("=" * 60)
print(f"eta = 0.3: RMSE = {rmse_03:.6f}")
print(f"eta = 0.1: RMSE = {rmse_01:.6f}")
print(f"\nDifference: {abs(rmse_03 - rmse_01):.6f}")

if rmse_03 < rmse_01:
    print("\nBest eta: 0.3 (lower RMSE)")
    best_eta = 0.3
    best_rmse = rmse_03
elif rmse_01 < rmse_03:
    print("\nBest eta: 0.1 (lower RMSE)")
    best_eta = 0.1
    best_rmse = rmse_01
else:
    print("\nBoth eta values give equal RMSE")
    best_eta = "Both equal"
    best_rmse = rmse_03

print(f"\nAnswer: eta = {best_eta} leads to the best RMSE score ({best_rmse:.6f})")
```

```python
# Visualize the comparison
etas = [0.3, 0.1]
rmses = [rmse_03, rmse_01]

plt.figure(figsize=(8, 6))
bars = plt.bar(['eta=0.3', 'eta=0.1'], rmses, color=['steelblue', 'coral'], alpha=0.7)
plt.ylabel('RMSE', fontsize=12)
plt.title('XGBoost RMSE Comparison: eta=0.3 vs eta=0.1', fontsize=14)
plt.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, rmse in zip(bars, rmses):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{rmse:.6f}',
             ha='center', va='bottom', fontsize=11)

plt.tight_layout()
plt.show()
```

### Answer to Question 6

**Answer: 0.1**

**Explanation**:
The comparison shows:
- eta = 0.3: RMSE = **0.450178**
- eta = 0.1: RMSE = **0.426228**

The model with **eta = 0.1** achieves a lower RMSE (0.426228) compared to eta = 0.3 (0.450178). This represents a **5.3% improvement** in RMSE.

The lower learning rate (eta = 0.1) allows the model to make smaller, more careful adjustments at each boosting round. Over 100 rounds, these smaller steps accumulate to find a better solution than the larger steps taken with eta = 0.3. The higher learning rate (0.3) likely overshoots optimal values, preventing the model from converging to the best possible solution within 100 rounds.

**Answer: 0.1** leads to the best RMSE score.

### Insights for Question 6

The `eta` parameter (learning rate) controls how much each tree contributes to the final prediction. A lower eta (0.1) makes the model learn more slowly but potentially more accurately, while a higher eta (0.3) makes it learn faster but might overshoot optimal values. The comparison shows which learning rate performs better for this specific dataset and number of boosting rounds (100). This result demonstrates the importance of tuning the learning rate in gradient boosting - even with the same number of boosting rounds, the learning rate can significantly impact final model performance. In practice, you might need more rounds with eta=0.1, but given 100 rounds, the slower learning rate produces better results.
